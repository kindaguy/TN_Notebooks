## Arrays and matrices

```julia
# Create a zero array of complex numbers with 3 elements
cArrZero = zeros(ComplexF64, 3)
# Create a array of complex numbers with 3 elements
cArr = ComplexF64[1+2im, 3+4im, 5+6im]
```

```julia
cArrZero
```

A complex valued matrix can be created as:

```julia
# Create a zero matrix of complex numbers with 2 rows and 2 columns
cMatZero = zeros(ComplexF64, 2, 2)
# Create a matrix of complex numbers with 2 rows and 2 columns
cMat = ComplexF64[1+2im 3+4im; 5+6im 7+8im]
```

### Trace

The trace of a matrix is computed as:

```julia
tr(cMat)
```

which is the same as summing the elements on the diagonal of `cMat`

```julia
tr(cMat) == sum(diag(cMat))
```

The trace is a linear operation:

```julia
β = 1.0 - 1.0im
γ = 2.0 + 3.0im
cMat2 = ComplexF64[1 2; 3 4]
```

```julia
tr(β * cMat + γ * cMat2) == β * tr(cMat) + γ * tr(cMat2)
```

### Eigensystem

Solving the eigensystem of `cMat`: `eigen` returns a couple `(eigenvaluse,eigenvectors)`, with the eigenvectors the columns of a matrix

```julia
eigvls,eigvcs = eigen(cMat)
```

```julia
cMat * eigvcs[:,1] ==  eigvcs[:,1]*eigvls[1]
```

```julia
≈
```

...remember that finite precision is a tricky business...

```julia
cMat * eigvcs[:,1] ≈ eigvcs[:,1]*eigvls[1]
```

```julia
cMat * eigvcs[:,2] ≈ eigvcs[:,2]*eigvls[2]
```

### Singular Value Decomposition

Singular Value Decomposition (SVD) of a matrix:

```julia
uu,ss,vv = svd(cMat)
```

```julia
uu * Diagonal(ss) * vv' ≈ cMat
```

```julia
function orthonormalize(v::Vector{ComplexF64})
    return v / norm(v)
end
```

```julia
norm(orthonormalize(cArr))
```

```julia
function orthonormalize(v::Vector{Vector{ComplexF64}})
    #Create an empty vector of vectors of complex numbers
    res = Vector{Vector{ComplexF64}}()
    #Check if the input vector is empty
    if(length(v) == 0)
        println("The input vector is empty, returning the empty vector")
        return v
    else
        size = length(v)
        appo = v[1]/norm(v[1])
        for i in 1:size
            push!(res, appo)
            if i < size
                appo = v[i+1] - sum(res[j]' * v[i+1] * res[j] for j in 1:i)
                appo = appo / norm(appo)
            end

        end
        return res
    end
end
```

```julia
basis = orthonormalize([cArr, cArr2])
```

This is an example of Julia's multiple dispatch: Julia determines which function to actually invoke by checking the type of the input parameters.
