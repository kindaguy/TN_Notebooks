Especially, we will see how MPS *can* be more compact than an explicit representation of the quantum state as an element of the Hilbert space, and how they store information about the quantum state.

## Exercise 1: The curse of dimensionality

In this exercise, we want to get a feeling for the "*curse of dimensionality*", and how tensor networks (TNs) offer a work around.

We will work with the GHZ state that was introduced in the lecture
$$|\text{GHZ}\rangle = \frac{1}{\sqrt{2}}\left(|0\rangle^{\otimes N} + |0\rangle^{\otimes N}\right) \ ,$$
where $|0\rangle = (1\ 0)^T$ and $|1\rangle = (0\ 1)^T$.

### **1) Creating the GHZ state**

**1.1)** Define the single-particle states `ket0` and `ket1`

**1.2)** Write a recursive function `tensorpower` that takes as an input a single-particle state $|\phi\rangle$ and an integer $N$ and returns a $N$-body state $|\phi\rangle^{\otimes N}$

**1.3)** Write a function `GHZstate` that takes as an input an integer $N$ and returns the GHZ state

**1.4)** Call `GHZstate` for $N = 3, 5, 10, 100,$ etc. What do you notice?

````julia
ket0 = [1.0, 0];
ket1 = [0, 1.0];
````

````julia
function tensorpower(state::Vector, m::Int64, state2::Vector=[])
    if m<=0
        ArgumentError("m must be positive")
    elseif m==1
        return state2
    else
        state2 = isempty(state2) ? kron(state, state) : kron(state2, state)
        tensorpower(state, m-1, state2)
    end
end
````

````julia
function GHZstate(N::Int)
    return (tensorpower(ket0, N) + tensorpower(ket1,N))/sqrt(2)
end
````

````julia
GHZstate(100) # Kernel dies
````

## Exercise 2: Bond dimension, Singular Value Spectrum, and von Neumann entropy

In this exercise, we will investigate the relation (of the lack thereof) between the bond dimension, the singular values and the von Neumann entropy of a quantum state.

We recall that the (binary) von Neumann entropy of a reduced state $\rho_{[1:\ell]}$ defined by the bipartition (1, $\ldots$, $\ell$ | $\ell+1, \ldots, N$) is given by

$$ S(\rho_{[1:\ell]}) = -\sum_{n=1}^{D_\ell} s_n^2 \log_2(s_n^2)\ ,$$

where $D_\ell$ is the bond dimension of bond $\ell$ and $s_n$ are the singular values (Schmidt coefficients) at that bond.

### 1) GHZ state

**1.1)** The function `orthogonalize` takes as an input a MPS $A$ and an integer $k$ labelling a site, and returns a MPS in a site-centered mixed canonical gauge with the orthogonality center at site $k$. 
Write a function `entropy` that takes as an input a MPS and returns the von Neumann entropy of all the bipartitions of the state.

**1.2)** Using the MPS representation of the GHZ state that you used in Exercise 1, compute and plot the von Neumann entropy of all its bipartitions.

**1.3)** What is the maximal possible value of the von Neumann entropy for a bond dimension $D_\ell$?

**1.4)** What are the bond dimensions of the MPS representation of the GHZ state?

**1.5)** Interpret the result you obtained in **2)**.

````julia
function entropy(mps)
    N = length(mps)
    spectrum = Float64[]
    for b=1:N-1
        mps = orthogonalize(mps, b)
        U,S,V = svd(mps[b], (linkinds(mps, b-1)..., siteinds(mps, b)...))
        SvN = 0.0
        for n=1:dim(S, 1)
          p = S[n,n]^2
          SvN += -p * log2(p)
        end
        push!(spectrum, SvN)
    end
    return spectrum
end
````

````julia
plot(entropy(ψ), label=:none, xlabel=L"bond index $\ell$", ylabel="Entropy", lw=4)
hline!([log2(2)], ls=:dash, lw=4, label=L"\log_2(2)", ylims=(0,2))
````

### 2) Random MPS



````julia
randmps = random_mps(sites, linkdims=256)#Int64.(volumelaw))
normalize!(randmps)
Nhalf = Int64(0.5N)
mixed_rand = orthogonalize(randmps, Nhalf+1)
_,Srand,_ = svd(mixed_rand[Nhalf+1], (linkinds(mixed_rand, Nhalf)..., siteinds(mixed_rand, Nhalf+1)...));
````

````julia
slope = (Srand[end] - Srand[1])/(size(Srand)[1]-1)
intercept = Srand[1]
linearfit = [slope*(n-1) + intercept for n=1:size(Srand)[1]];
````

````julia
plot(linearfit, lw=4, label="Linear Fit")
plot!(diag(matrix(Srand)),m=:circle, lw=4, xlabel=L"n", ylabel=L"s_n", label="Random MPS")
````

````julia
plot(entropy(randmps), label=:none, xlabel=L"bond index $\ell$", ylabel="Entropy", lw=4)
hline!([8], ls=:dash, lw=4, label=L"\log_2(256)")
````

````julia
R = abs.(randn(fill(2, 10)...))
mpsR = MPS(R, siteinds("Qubit",10);)
normalize!(mpsR)
plot(entropy(mpsR), label=:none, xlabel=L"bond index $\ell$", ylabel="Entropy", lw=4)
hline!([5], ls=:dash, lw=4, label=L"\log_2(32)")
````

````julia
mixed_randR = orthogonalize(mpsR, 5)
_,SrandR,_ = svd(mixed_randR[5], (linkinds(mixed_randR, 4)..., siteinds(mixed_randR, 5)...));
plot(diag(matrix(SrandR)),m=:circle, lw=4, xlabel=L"n", ylabel=L"s_n", label="Random MPS")
````
